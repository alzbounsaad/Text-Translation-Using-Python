```python
# Load library
from nltk.corpus import stopwords  # import stopwords from NLTK (Natural Languages ToolKit) to use english stopwords
from googletrans import Translator  # import Translator from googletrans module to call google translate API
import nltk
from tkinter import *
```

```python
nltk.download('stopwords')  # download the last version of english stopwords
stop_words = stopwords.words('english')  # save the english stopwords in a list
translator = Translator()  # call google API
```

```python
def translate():
    # read data to translate
    title_file = open(
        "Hellboy.2019.HC.HDRip.XviD.AC3-EVO.srt",
        encoding='utf-8')
    title = title_file.readline()

    # write data after translate to a file
    out_file = open("translated_data.srt", encoding='utf-8', mode='w')

    source_lang = scr_Entry.get()
    destination_lang = des_Entry.get()

    while title:
        if title == '\n':  # check if the line that reads from the file is empty or not
            out_file.write('\n')  # write empty line to the output file if the line is empty

        else:
            if title[1].isdigit():  # check if the the line that reads from the file is start with digit or not
                out_file.write(title.strip())  # write the line without make any changes on it if start with digit

            else:
                data_tokenize = title.lower().split()  # tokenize the received sentence in order to translate it
                print('Tokenize data: ', data_tokenize)

                for word in data_tokenize:  # translate the words one by one
                    if word not in stop_words:
                        if data_tokenize.index(word) == 0:
                            res = translator.translate(word, src=source_lang, dest=destination_lang)

                            print(word.capitalize(), '(', res.text, ')', end=' ')
                            out_file.write(word.capitalize() + ' ' + '(' + res.text + ')' + ' ')

                        else:
                            res = translator.translate(word, src=source_lang, dest=destination_lang)

                            print(word.capitalize(), '(', res.text, ') '.strip(), end=' ')
                            out_file.write(word + ' ' + '(' + res.text + ')' + ' ')

                    else:
                        if data_tokenize.index(word) == 0:
                            print(word.capitalize().strip(), end=' ')
                            out_file.write(word.capitalize() + ' ')
                        else:
                            print(word.strip(), end=' ')
                            out_file.write(word + ' ')
        print('\n')
        out_file.write('\n')
        title = title_file.readline()  # reads the next line from the file
    from tkinter import messagebox

    messagebox.showinfo("Message", "Translate finished")
    out_file.close()  # close the output file after write translated text to it
```
